# When Machine Learning is used.
### 1. Explicit rules are hard to write
Some problems can’t be solved by simply writing if-else conditions.
- Example: Email spam detection — spammers constantly change wording, so fixed rules fail.

### 2. Too many possibilities to code manually
Some problems have such a large variety of possibilities that writing all cases is impossible.
- Example: Object detection in images — the same object can appear in endless shapes, sizes, and colors.

### 3. Data mining (When patterns are hidden and complex)
Sometimes the relationship between inputs and outputs isn’t obvious, even after looking at the data.
- Example: By reading an email, it might not be clear what specific keywords indicate spam. Machine learning can analyze large amounts of data, find hidden patterns, and create prediction models.

## Unsupervised Learning

<img src="../img/ML3.png" width ="900">

### 1. Supervised Learning
Supervised learning is a type of machine learning where the model is trained under supervision — just like a child learns under the guidance of a teacher. The data used for training is labeled, meaning we already know the correct answers. The model learns from these examples, and after enough training, it can make predictions on new, unseen data on its own.

### 2. Unsupervised Learning
Unsupervised learning, on the other hand, has no supervision — there are no labeled answers. The model tries to find hidden patterns or relationships in the data by itself, like grouping similar customers or identifying patterns in sales. Algorithms like K-Means Clustering and PCA are commonly used for this.

### 3. Semi-Supervised Learning

> **Note:** Labeling requires a lot of resources

In semi-supervised learning, we label only a few classes, and the model automatically labels the rest. For example, in Google Photos, the system detects and labels people’s faces automatically.

### 4. Reinforcement Learning
In reinforcement learning, there’s no training data provided to the model. It starts learning from scratch through trial and error. For example, a self-driving car learns by interacting with its environment. In reinforcement learning, the model is called an agent.”

## Batch (Offline) vs Online Machine Learning

### 1. Batch Learning
- Conventional method where the **entire dataset** is used to train the model at once (no incremental training).

- **Problem:** The model is static. Business and data evolve over time, but batch models require retraining with the **entire dataset offline** and redeploying it to the server.

- **Disadvantages:**

  1. Potential **loss of data** if updates are delayed.

  2. **Hardware limitations** (e.g., remote locations, army devices, satellites with no connectivity).

  3. **Availability issues** — retraining takes time.

- **Use Case:** Suitable for small-to-moderate datasets or when changes in data are infrequent.



### 2. Online Learning

- The model **learns incrementally**, continuously updating as new data arrives.

- **Advantages:**

  - Processes only a **small batch of data at a time**, reducing memory usage.

  - Can adapt to **data drift** (changes in patterns over time).

- **Example:** YouTube recommendation feed, stock price prediction.

- **Implementation Libraries:**

  1. scikit-learn (`SGDRegressor.partial_fit()`)

  2. River library

  3. Vowpal Wabbit



### 3. When to Use Batch vs Online

- **Batch:** Data is relatively static, retraining infrequently is acceptable.

- **Online:** Concept drift occurs frequently; data size is huge; low latency or cost-efficient updates are needed.



### 4. Learning Rate in Online Learning

- Determines **how much the model weighs new data versus old knowledge**.

- Critical for sequential data and **out-of-core learning** (datasets too large to fit in memory at once).



### 5. Disadvantages / Risks of Online Learning

1. **Complexity:** Requires careful setup; not trivial to implement perfectly.

2. **Risk of bias:** If new data is biased, the model can misbehave.



**Mitigation Strategies:**  

- Continuous or periodic monitoring  

- Anomaly detection  

- Rollback mechanisms in case of model failure

| Feature                 | Batch (Offline)                                                                        | Online (Incremental)                                                               |
| ----------------------- | -------------------------------------------------------------------------------------- | ---------------------------------------------------------------------------------- |
| **Complexity**          | Low                                                                                    | Dynamic / Higher                                                                   |
| **Computational Power** | High, used occasionally for full retraining                                            | Continuous, small incremental updates                                              |
| **Use in Production**   | Easier to implement                                                                    | More challenging; needs monitoring and careful updates                             |
| **Applications**        | Traditional ML tasks with static data (e.g., image classification, payroll prediction) | Real-time tasks, streaming data, recommendations (e.g., YouTube, stock prediction) |
| **Tools / Libraries**   | scikit-learn (full training), TensorFlow, PyTorch                                      | scikit-learn (`partial_fit()`), River, Vowpal Wabbit                               |

## Instance Based Learning vs Model Based Learning

<img src="../img/ML5.png" width="650">

### 1. Instance-Based (Lazy)
- **Definition:** Memorizes training data; predicts by comparing new instances to stored ones.
- **Computation:** Minimal during training, heavy during prediction.
- **Memory:** High (stores all data).
- **Pros:** Adapts easily to new data; no assumptions about data distribution.
- **Cons:** Slow for large datasets; high memory usage.
- **Example:** k-NN
- **Analogy:** “Learn by remembering past examples.”

### 2. Model-Based (Eager)
- **Definition:** Learns a general model/pattern from training data for predictions.
- **Computation:** Heavy during training, fast during prediction.
- **Memory:** Low (stores model parameters, not all data).
- **Pros:** Fast predictions, can generalize well if model fits data.
- **Cons:** Needs retraining if data changes; may make wrong assumptions (bias).
- **Example:** Linear Regression, Decision Tree, Neural Network
- **Analogy:** “Learn the rules first, then apply to new situations.”

## Challenges in Machine Learning

1. **Data Collection**
   - Collecting sufficient and relevant data is often difficult.

2. **Insufficient / Unlabeled Data**
   - Labeling data is tedious and time-consuming.

3. **Non-Representative Data**
   - Data may not cover all cases, leading to sampling bias or noise.
   - Example: World Cup data collected only from the Indian team; other teams are missing.

4. **Poor Quality Data**
   - Outliers, 
   - missing values, 
   - inconsistent formats (different foramt).

5. **Irrelevant Features**
   - Garbage in → Garbage out.
   - Example: Predicting health using features like age, weight, height — location may be irrelevant.

6. **Overfitting**
   - Model performs very well on training data but poorly on test data.
   - Occurs when model learns noise and outliers.
   - **Note:** 100% training accuracy usually indicates overfitting.

7. **Underfitting**
   - When Model is too simple to capture patterns in data.
   - Example: Fitting a straight line, when there is a curve.

8. **Software Integration**
   - Integrating ML models into existing systems can be challenging.

9. **Offline (Batch) Learning & Deployment**
   - Constant deployment difficulty.
   - High cost due to retraining on full datasets.

10. **Cost Involvement**
    - Server and infrastructure costs; hidden charges.

## ML Development Life Cycle (MLDLC / MLDC)

### 1. **Frame the Problem**
   - Define the problem to be solved.
   - Understand the customer, cost, team, and end product.
   - Decide if the model is supervised or unsupervised.
   - Determine if offline/batch or online learning.
   - Select the algorithm type and required data.

### 2. **Data Gathering**
   - Sources: CSV, API, Web scraping, Database warehousing.
   - **Data Warehousing:** Extract production data safely into a separate warehouse for ML use.

### 3. **Data Preprocessing**
   - Remove duplicates.
   - Handle missing values.
   - Detect and treat outliers.
   - Scale and standardize features.

### 4. **Exploratory Data Analysis (EDA)**
   - Visualize data.
   - Univariate, Bivariate, and Multivariate analysis.
   - Detect imbalance and balance the data.
   - Detect outliers.

### 5. **Feature Engineering and Selection**
   - Create new features, update existing ones.
   - Select impactful features to reduce dimensionality and training time.

### 6. **Model Training, Evaluation & Selection**
   - **Training:** Use different algorithms to train models.
   - **Evaluation:** Compare performance using metrics like Accuracy, MSE, F1-score.
   - **Selection:** Choose the best algorithm and tune hyperparameters.
   - **Ensemble Learning:** Combine multiple models for improved performance.

### 7. **Model Deployment**
   - Deploy the model to production servers.
   - Ensure integration with applications and APIs.

<img src="../img/ML8.png" width="900">


### 8. **Testing**
   - **Beta Testing:** Provide features to select users for feedback.
   - **A/B Testing:** Compare variations; if problems occur, iterate previous steps.

### 9. **Optimization**
   - Optimize for all users on the server.
   - Implement backups, load balancing.
   - Decide model update frequency to prevent “model rotting.”
   - Automate processes for efficiency and cost-effectiveness.
